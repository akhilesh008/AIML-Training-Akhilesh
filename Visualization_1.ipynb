```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)
```

```python
x = np.random.random(100) #generate 100 real random numbers in the range 0 to 100
y = np.random.normal(size=100) # generate 100 random numbers in normal distribution
```

## Distribution of data

```python
plt.hist(x, bins=20)
```

```python
plt.hist(y,bins=20, color='red')
```

```python
sns.histplot(y, bins=20,kde=True) # Kernel data estimation
```

```python
sns.histplot(x,bins=20, kde=True)
```

```python
sns.kdeplot(y, color='magenta', fill=True) # if we want only curve we can use kdeplot; fill is to fill the area
```

```python
# to plot two data distribution in single graph
sns.kdeplot(x,color='red',fill=True)
sns.kdeplot(y,color='green',fill=True)
```

```python
# to arrange plots horizontally
plt.subplot(1,2,1)
sns.kdeplot(x,color='red')
plt.subplot(1,2,2)
sns.kdeplot(y, color='blue')
```

```python
# to arrange plots vertically
plt.subplot(2,1,1)
sns.kdeplot(x,color='red')
plt.subplot(2,1,2)
sns.kdeplot(y, color='blue')
```

### Now data is available , so draw plots

```python
sns.histplot(tips_df.total_bill, color='green', bins=20) # to plot details of totls bills column
```

```python
sns.histplot(tips_df.tip, color='red',bins=20)
```

```python
sns.histplot(tips_df.tip/tips_df.total_bill, color='orange')
```

```python
tips_df[tips_df.tip/tips_df.total_bill >0.5] # to know the data for the person who has given tip > 50%
```

```python
tips_df[tips_df.tip/tips_df.total_bill >0.4] # to know the data for the person who has given tip > 40%
```

```python
tips_df[tips_df['day']=='Sun']['day'].count() # to get number of sundays in the dataset
```

```python
days = tips_df.day.unique()
day_data = {}

for each_day in days:
  day_data[each_day]= [tips_df[tips_df['day']==each_day]['day'].count()] # make a list of values so that it can be converted to dataframe

print(day_data)
```

```python
day_df = pd.DataFrame(day_data) # convert the list into pandas dataframe
sns.barplot(day_df)
```

```python
# group by in pandas
tips_df.groupby('day')['tip'].count()
```

## Using series of pandas

```python
tips_series = tips_df.groupby('day')['day'].count()
sns.barplot(x=tips_series.index, y=tips_series.values) # makes it simple to plot
```

```python
tips_series = tips_df.groupby('sex')['day'].count()
sns.barplot(x=tips_series.index, y=tips_series.values) # makes it simple to plot
```

#### Using relational plots

```python
# basic plot is rel plot
sns.relplot(data=tips_df, x='total_bill',y='tip')
```

```python
# regression plot
sns.regplot(data=tips_df, x='total_bill',y='tip')
```

```python
# linear model plot
sns.lmplot(data=tips_df, x='total_bill',y='tip')
```

```python
# linear model plot
sns.lmplot(data=tips_df[tips_df.sex=='female'], x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips_df,x='total_bill', y='tip', hue='day')
```

```python
sns.lmplot(data=tips_df,x='total_bill', y='tip', hue='day', col='smoker')
```

```python
sns.lmplot(data=tips_df,x='total_bill', y='tip', hue='day', col='smoker', row='sex')
```

```python
# pair plot to get relatoinship between all attributes
# gives relationship between numerical values
sns.pairplot(tips_df)
```

## Donut charts
### Composition of data  

```python
tips_series = tips_df.groupby('day')['day'].count()
plt.pie(tips_series, labels=tips_series.index, autopct='%0.2f%%', shadow=True, explode=[1,1,0,0])
```

## Donut chart is better than pychart

```python
# to use pychart into donut chart
plt.pie(tips_series, labels=tips_series.index, autopct='%0.2f%%', shadow=True, wedgeprops=dict(width=0.25)) # add wedge to convert to donut chart
```

```python
plt.pie(tips_series, labels=tips_series.index, autopct='%0.2f%%', shadow=True, wedgeprops=dict(width=0.25), explode=[1,0,0,0])
```
